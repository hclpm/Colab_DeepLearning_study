```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
```

```python
x = np.array([1,3,5,6,10], dtype=np.float32)
y = np.array([0.5,1,2,1.8,3], dtype=np.float32)
```

```python
print(x)
print(y)
```

```python
model = keras.Sequential([
    # unit= 레이어 내의 노드 개수, input_shape= 노드에 입력되는 데이터 수, activation=노드의 작동 방법
    layers.Dense(units=1, input_shape=[1], activation="linear")
])

init_w, init_b = model.weights
print(init_w, init_b)
# >> <tf.Variable 'dense/kernel:0' shape=(1, 1) dtype=float32, numpy=array([[0.49368656]], dtype=float32)> <tf.Variable 'dense/bias:0' shape=(1,) dtype=float32, numpy=array([0.], dtype=float32)>
# 초기 가중치: 0.49368656, 초기 편향: 0. 으로 설정된 것을 확인 할 수 있음
```

```python
plt.scatter(x, y)
plt.plot(x, init_w[0][0] * x + init_b[0], color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title("init Linear Regression")
plt.show()
```

```python
# 모델 학습 과정

# 손실함수 최적화에 쓸 최적화 알고리즘 지정(SGD: Stochastic Gradient Algorithm) + 학습률 지정
sgd = keras.optimizers.SGD(learning_rate = 0.001)

# 모델 컴파일(최적화 기술: SGD, 손실함수 형태: MSE)
model.compile(optimizer=sgd, loss='mean_squared_error')

# 모델 학습(x, y, batch_size=한번에 학습할 데이터의 개수, epoch= 전체 데이터 학습 반복횟수)
history = model.fit(x, y, batch_size=1, epochs=10)

w,b = model.layers[0].get_weights()[0][0][0], model.layers[0].get_weights()[0][0]

print(f'\n\nTrained W : {w}, Trained b: {b}\n')

plt.scatter(x, y)
plt.plot(x, w * x + b, color = 'red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```
